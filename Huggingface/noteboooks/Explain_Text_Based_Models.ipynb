```python
%load_ext autoreload
%autoreload 2

from Huggingface.utils import HuggingfaceDatasetManager, TransformersModelTypeEnum, DatasetTypeEnum, ModelManager, barplot_token_shap_values, get_most_important_n_tokens
```

## Text Based Models

```python
# All models share the same dataset, so we load it for all of them.
dataset_manager = HuggingfaceDatasetManager(dataset_type=DatasetTypeEnum.KAGGLE_FAKE_NEWS)
```

### 1. bert-base-uncased

```python
bert_base_uncased_model_type = TransformersModelTypeEnum.EZ_BERT_BASE_UNCASED_FAKE_NEWS
bert_base_uncased_model_manager = ModelManager(bert_base_uncased_model_type)
```

```python
labels, samples = dataset_manager.fetch_random_samples(model_type=bert_base_uncased_model_type)
shap_values_dict_list_BBU, explainer_BBU = bert_base_uncased_model_manager.explain_texts(samples, visualize=False)
```

```python
# analyze the first sample
tokens_BBU, shap_values_BBU = get_most_important_n_tokens(shap_values_dict_list_BBU[0], labels.iloc[0], n=10)
barplot_token_shap_values(tokens_BBU, shap_values_BBU)
```

### 2. distilroberta-base

```python
distilroberta_base_model_type = TransformersModelTypeEnum.GA_DISTIL_ROBERTA_BASE_FINETUNED_FAKE_NEWS
distilroberta_base_model_manager = ModelManager(distilroberta_base_model_type)
```

```python
# labels, samples = dataset_manager.fetch_latest_samples(model_type=distilroberta_base_model_type)
# first 2 models share the same preprocessing behavior so we don't need to fetch the samples
shap_values_dict_list_DRB, explainer_DRB = distilroberta_base_model_manager.explain_texts(samples, visualize=False)
```

```python
# analyze the first sample
tokens_DRB, shap_values_DRB = get_most_important_n_tokens(shap_values_dict_list_DRB[0], labels.iloc[0], n=10)
barplot_token_shap_values(tokens_DRB, shap_values_DRB)
```

### 3. roberta_base

```python
roberta_base_model_type = TransformersModelTypeEnum.HB_ROBERTA_FAKE_NEWS
roberta_base_model_manager = ModelManager(roberta_base_model_type)
```

```python
labels_RB, samples_RB = dataset_manager.fetch_latest_samples(model_type=roberta_base_model_type)
shap_values_dict_list_RB, explainer_RB = roberta_base_model_manager.explain_texts(samples_RB, visualize=False)
```

```python
# analyze the first sample
tokens_RB, shap_values_RB = get_most_important_n_tokens(shap_values_dict_list_RB[0], labels_RB.iloc[0], n=10)
barplot_token_shap_values(tokens_RB, shap_values_RB)
```
