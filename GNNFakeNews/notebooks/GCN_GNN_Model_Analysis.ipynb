```python
from GNNFakeNews.utils.helpers import GNNModelTypeEnum
from torch_geometric.nn import GNNExplainer
from util import run_model, GNNModelExplainer
import matplotlib.pyplot as plt
from GNNFakeNews.utils.helpers import HparamFactory
from GNNFakeNews.utils.enums import GNNDatasetTypeEnum, GNNFeatureTypeEnum
```

# 1.1. GCN GNN

```python
model_type = GNNModelTypeEnum.GCN_GNN
```

## DATASET TYPE = POLITIFACT, FEATURE = BERT

```python
gcn_model_politifact_bert, dataset_manager_politifact_bert = run_model(model_type, return_dataset_manager=True, local_load=True)

sample_data = dataset_manager_politifact_bert.get_train_samples()[0].to(gcn_model_politifact_bert.m_args.device)

gcn_model_politifact_explainer = GNNModelExplainer(gcn_model_politifact_bert, sample_data)
```

## DATASET TYPE = POLITIFACT, FEATURE = SPACY

```python
gcn_model_politifact_spacy, dataset_manager_politifact_spacy = run_model(model_type, return_dataset_manager=True, local_load=True, hyperparams=HparamFactory(model_type, feature=GNNFeatureTypeEnum.SPACY))

sample_data = dataset_manager_politifact_spacy.get_train_samples()[0].to(gcn_model_politifact_spacy.m_args.device)
gcn_model_politifact_explainer = GNNModelExplainer(gcn_model_politifact_spacy, sample_data)
```

## DATASET TYPE = POLITIFACT, FEATURE = PROFILE

```python
gcn_model_politifact_profile, dataset_manager_politifact_profile = run_model(model_type, return_dataset_manager=True, local_load=True, hyperparams=HparamFactory(model_type, feature=GNNFeatureTypeEnum.PROFILE))

sample_data = dataset_manager_politifact_profile.get_train_samples()[0].to(gcn_model_politifact_profile.m_args.device)
gcn_model_politifact_explainer = GNNModelExplainer(gcn_model_politifact_profile, sample_data)
```
